```python
import cloudscraper
from bs4 import BeautifulSoup
import requests
import re

url = "https://www.lazada.com.my/products/graphic-card-zotac-gaming-nvidia-geforce-rtx-3080-amp-holo-i1734238850-s6652768621.html?spm=a2o4k.seller.list.1.6c4b7a79X3yOvN&mp=1"

scraper = cloudscraper.create_scraper()
text = scraper.get(url).text

soup = BeautifulSoup(text,'html5lib')

scraped_text = soup.get_text()

def re_text(scraped_text):
    if re.findall("Out of Stock", scraped_text) == 0:
        print("Stock Available")
    else:
        print("Out of Stock")
```
